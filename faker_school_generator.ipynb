```python
from collections import defaultdict
import numpy as np
import uuid

fake = Faker(['pt_BR'])

#class faker.providers.person.pt_BR.Provider(generator: Any)

fake_alunos = defaultdict(list)
fake_ocorrencias = defaultdict(list)
fake_atividades = defaultdict(list)
fake_financeiro = defaultdict(list)

ocorrencias_areas = ['Pátio', 'Cantina', 'Sala de aula', 'Quadra de Esportes', 'Banheiros','Laboratório','Biblioteca']
ocorrencias_tipo = ['Grave', 'Notificação', 'Moderada', 'Leve']
atividades_tipo = ['Prova', 'Esportiva', 'Trabalho de casa','Trabalho em grupo','Laboratorial']
atividade_extra = ['Curso Extra','Compra de Livros', 'Atividade Cultura']
genero_tipo = ['Masculino','Feminino','Não-binário']
etnia_tipo = ['Indígena','Negro','Branco','Pardo','Mulato','Caboclo,','Cafuzo']
classe_matricula = ['A','B','C']
serie_matricula = ['1º','2º','3º']
status_matricula = ['Ativa','Inativa']
uf_endereco = ['SP','RJ','MG','BA','CE','PE','PI','PR','SC','MA','AM']
status_aula =['Presente','Falta']
disciplinas = ['Matemática','Português','Biologia','Química','Geografia','Informática','Física','Astronomia']
grau_parentesco = ['1º','2º','3º']
professores_nivelescolar = ['Superior Completo','Mestrado','Doutorado']
prioridades_tipo = ['Baixa','Média','Alta']
chamado_motivo = ['Problemas com acesso na escola','Problemas com acesso ao portal','Boleto sem baixa','Atualização cadastral','Financiamento escolar','Programa de inclusão','Isenção de taxas','Outros']
aula_nome = ['Normal','Conjunta','Aula de campo','Laboratorial']
pagExtra_tipo = ['Unidade Extra Curricular', 'Livro', 'Teatro', 'Museu', 'Evento']


#ALunos
for _ in range (1000):
    
    id_aluno = uuid.uuid4().int
    fake_alunos["IdAluno"].append(id_aluno)
    fake_alunos["PrimeiroNome"].append(fake.first_name()) 
    fake_alunos["Sobrenome"].append(fake.last_name())
    fake_alunos["Idade"].append(fake.random_int(14, 27))
    fake_alunos["Etnia"].append(np.random.choice(etnia_tipo))
    fake_alunos["Genero"].append(np.random.choice(genero_tipo))
    fake_alunos["UF"].append(np.random.choice(uf_endereco))
    fake_alunos["email"].append(fake.email())
     
    #fake_alunos["DataNasc"].append(np.random.choice(etnia_tipo))

#Ocorrencias
    for _ in range(10):
        id_ocorrencia = uuid.uuid4().int
        fake_ocorrencias['idOcorrencia'].append(id_ocorrencia)
        fake_ocorrencias['idAluno'].append(id_aluno)
        fake_ocorrencias['Area'].append(np.random.choice(ocorrencias_areas))
        fake_ocorrencias['Tipo'].append(np.random.choice(ocorrencias_tipo))

#Atividades
    quantidade_de_atividades = np.random.randint(1,7)
    for _ in range (quantidade_de_atividades):
        id_atividade = uuid.uuid4().int
        fake_atividades['idAtividade'].append(id_atividade)
        fake_atividades['idAluno'].append(id_aluno)
        fake_atividades['Tipo'].append(np.random.choice(atividades_tipo))
        fake_atividades['Data'].append(fake.date_between(start_date='-6m', end_date='today'))
        fake_atividades['Nota'].append(fake.random_int(40, 100))

#Financeiro    
    for _ in range (4):
        id_financeiro = uuid.uuid4().int
        fake_financeiro['idFinanceiro'].append(id_financeiro)
        fake_financeiro['idAluno'].append(id_aluno)
```

```python
import pandas as pd
pd.DataFrame(fake_alunos).to_csv("fake_alunos.csv", index=False)
pd.DataFrame(fake_ocorrencias).to_csv("fake_ocorrencias.csv", index=False)
pd.DataFrame(fake_atividades).to_csv("fake_atividades.csv", index=False)
pd.DataFrame(fake_financeiro).to_csv("fake_financeiro.csv", index=False)
```
